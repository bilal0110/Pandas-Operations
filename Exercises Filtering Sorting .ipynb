### Step 4. How many products cost more than $10.00? 

Use `str` attribute to remove the $ sign and convert the column to proper numeric type data before filtering.


```python
chipo['item_price']=[i.strip('$') if type(i)==str else i for i in chipo['item_price']]
```

```python
chipo['item_price']=chipo['item_price'].astype('float')
```

# Step 4.How many products cost more than $10.

```python
df=chipo[chipo['item_price']>=10.00]
print('We Have',df['item_name'].nunique(),'products which costs more than 10$')
```

### Step 5. What is the price of each item? 
print a data frame with only two columns `item_name` and `item_price`

```python
chipo.groupby(['item_name'])['item_price'].sum().sort_values(ascending=False).reset_index()
```

```python
chipo[['item_name','item_price']]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.groupby(['item_name'])['item_price'].max().reset_index()
```
