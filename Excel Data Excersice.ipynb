## Import some excel data (coalpublic2013.xlsx ) 

```python
df=pd.read_excel('coalpublic2013.xlsx')
```

##  Add summation to a row 

```python
print('Sum of Production',df['Production'].sum())
print('Sum of Labour Hours',df['Labor_Hours'].sum())
```

```python
df=df.append({'Production':df['Production'].sum(),'Labor_Hours':df['Labor_Hours'].sum()},ignore_index=True)
```

```python
df.tail(1)
```

## Create a subtotal of "Labor Hours" against MSHA ID

```python
df.groupby(['MSHA ID'])['Labor_Hours'].sum().reset_index()
```
