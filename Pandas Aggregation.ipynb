# Aggregations

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
df=pd.read_csv('Occupation.csv',encoding="latin",sep='|')
df
```

```python
print(df.info())
```

### Step 3. Assign it to a variable called users.

```python
users=df
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(['occupation'])['age'].mean().sort_values(ascending=False).reset_index().rename(columns={"age":"Min Age"})
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
df['gender'].replace({'M':1,'F':2},inplace=True)
```

```python
x=df.groupby(['occupation','gender'])['gender'].count()
y=df.groupby('occupation')['gender'].count()
x/y
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(['occupation'])['age'].agg([min,max]).reset_index().rename(columns={"min":"Min Age","max":"Max Age"})
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender'])['age'].mean().reset_index().rename(columns={"age":"Mean Age"})
```

### Step 8.  For each occupation present the percentage of women and men

```python
x=users.groupby(['occupation'])['gender'].count()
```

```python
y=users.groupby(['occupation','gender'])['gender'].count()
x/y
```
