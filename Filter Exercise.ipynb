## find out the alcohol consumption of any specific year from the world alcohol consumption dataset.

## find out the alcohol consumption details in the year '1987' or '1989' from the world alcohol consumption dataset

```python
df[(df['Year']==1987) | (df['Year']==1989)].head(5)
```

## find out the alcohol consumption details by the 'Americas' in the year '1985' from the world alcohol consumption dataset

```python
df[(df['WHO region']=='Americas') & (df['Year']==1985)]
```

## find out the alcohol consumption details in the year '1986' where WHO region is 'Western Pacific' and country is 'VietNam'

```python
df[(df['Year']==1986) & (df['WHO region']=='Western Pacific') & (df['Country']=='Viet Nam')]
```

## find out the alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas

```python
df[((df['Year']==1986) | (df['Year']==1989)) & (df['WHO region']=='Americas')]
```

## find out the 'WHO region, 'Country', 'Beverage Types' in the year '1986' or '1989' where WHO region is 'Americas' or 'Europe'

```python
df[((df['Year']==1986)  | (df['Year']==1989)) &  ((df['WHO region']=='Americas') | (df['WHO region']=='Europe'))]
```

## find out the records where consumption of beverages per person average >=5 and Beverage Types is Beer

```python
df[(df['Display Value']>=5) & (df['Beverage Types']=='Beer')]
```

## find out the records where consumption of beverages per person average >=4 and Beverage Types is Beer, Wine, Spirits

```python
df[(df['Display Value']>=4) & ((df['Beverage Types']=='Beer') | (df['Beverage Types']=='Wine') | (df['Beverage Types']=='Spirits'))]
```
